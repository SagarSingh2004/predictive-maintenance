# Predictive Maintenance using Machine Learning on Sensor-Based Industrial Data

---

## Problem Statement
The goal of this project is to build a machine learning model to predict whether a machine is likely to experience failure based on real-time sensor data. Predictive maintenance plays a critical role in modern industries by enabling early detection of potential failures, thereby reducing downtime, improving operational efficiency, and minimizing maintenance costs.

The dataset will be analyzed to understand relationships between different machine parameters, and a classification model will be developed to predict the Target variable, indicating machine failure.

---

## Objective
To develop and compare multiple classification models that can accurately predict whether a machine will fail (1) or not (0), using historical sensor and operational data.

---

## Dataset Overview

The dataset consists of:

- 10000 records (rows)
- 9 features (columns)

Each record represents the operational condition of a machine at a given point in time.

---

# Data Dictionary
| Column Name              | Description                                           | Data Type              |
|--------------------------|-------------------------------------------------------|------------------------|
| UDI                      | Unique identifier for each machine record             | Integer (ID)           |
| Product ID               | Identifier for machine product category               | String (Alphanumeric)  |
| Type                     | Machine type: Low (L), Medium (M), High (H)           | Categorical            |
| Air temperature [K]      | Temperature of surrounding air (in Kelvin)            | Float (Continuous)     |
| Process temperature [K]  | Temperature during manufacturing process (Kelvin)     | Float (Continuous)     |
| Rotational speed [rpm]   | Machine rotational speed (revolutions per minute)     | Integer (Continuous)   |
| Torque [Nm]              | Torque generated by machine (Newton-meters)           | Float (Continuous)     |
| Tool wear [min]          | Tool usage duration (in minutes)                      | Integer (Continuous)   |
| Target                   | Machine failure indicator (0 = No Failure, 1 = Failure) | Binary (0/1)         |
