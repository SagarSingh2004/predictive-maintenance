# <center>Data Understanding</center>

```python
# Importing the required Libraries
import pandas as pd
```

```python
# Loading the dataset
df = pd.read_csv("data/predictive_maintenance.csv")
```

```python
# Checking the shape of the dataset
print(f"The Shape of the dataset is {df.shape}")
```

```python
# Checking the first five rows
df.head()
```

```python
# Checking the column names
df.columns
```

```python
# Cleaning the column names
df.columns = (
    df.columns
    .str.strip()                          # remove leading and trailing spaces
    .str.lower()                          # convert to lowercase
    .str.replace(' ', '_')                # replace spaces with _
    .str.replace('[^a-zA-Z0-9_]', '', regex=True)  # remove special characters
)

# Checking the updated column names
print(df.columns)
```

```python
# Checking the summary of the dataset
df.info()
```

### Insights

- The dataset consists of 4 integer features, 3 float features, 2 categorical (object) features.

This indicates a mix of numerical and categorical data.

- Most of the features are numerical (temperature, speed, torque, tool wear), which are directly usable for machine learning models after scaling.

- The features type and product_id are categorical and will require encoding before being used in machine learning models.

- The target variable target is of integer type and represents a binary classification problem (0: No Failure, 1: Failure).

- All columns have 10,000 non-null values, indicating that there are no missing values in the dataset.

- The dataset appears to be clean and well-structured, requiring minimal preprocessing.

```python
# Checking the statistical summary of numerical columns
df.describe()
```

### Insights
- The dataset contains 10,000 records with no missing values. The target variable is highly imbalanced, with only ~3.39% failure cases.
- Temperature features show low variance, indicating stable operating conditions, while rotational speed, torque, and tool wear exhibit significant variation and are likely important predictors.
- The UDI column appears to be an identifier and does not contribute to prediction. We will drop it

```python
# Checking the statistical Summary of Categorical Columns
df.describe(exclude='number')
```

### Insights
- The product_id column contains unique values for each record and does not contribute to prediction, hence it will be dropped.
- The type column is a categorical feature with three categories, with an imbalanced distribution where type L dominates with 60% of values. This feature will be retained and encoded during preprocessing.

### Insights
